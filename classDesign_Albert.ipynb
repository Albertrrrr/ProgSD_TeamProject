Introduction to Vehicle Class

属性 Attributes 
ID （主键)     Int(10)         区分不同的单车
bikeType      String          电动、自行车（建议用子类复写）
price         float           价格（建议用子类复写）
location      String          记录单车的当前位置
status        String          表示单车的状态（可用，维修）
isRented      bool            表示单车的租用状态（租用，可用）
isLocked      bool            表示单车的锁车状态（上锁，未上锁）
batteryStatus int(10)         记录电池的充电状态
renter        int(10)         记录租用者的信息

方法 Methods
init()                  初始化方法         创建单车对象时初始化属性，设置ID、位置、状态等初始值。
lock()                  锁定单车           锁定单车
unlock()                解锁单车           解锁单车
rent()                  租用单车           将单车分配给租用者，记录租用开始时间，改变租用属性
returnBike()            归还代车           将单车还回，记录租用结束时间和位置，计算租用费用
reportMaintenance()     保修              报告单车需要维修，将状态设置为维修中
completeMaintenance()   维修完成           维修完成后将单车状态设置为可用
vehicleDetails()        详细信息           用list的方式返回 车辆id 价格 当前状态 电池状态

Introduction to Order Class 
属性 Attributes 
orderID （主键)     Int(10)         订单唯一标识号，区分不同订单
renter             Int(10)         与class customer进行耦合 拿到租用者的信息
bike               Int(10)         与Class Vehicle进行耦合  拿到车的id
startTime          String          订单租用开始世界
endTime            String          订单租用结束时间
startLocation      String          订单开始位置
endLocation        String          订单结束位置
createTime         String          订单创建时间
finishTIme         String          订单关闭时间
cost               float           总花费
isPaid             bool            是否支付
status             String          表示订单的状态，如已完成、进行中、已取消

方法 Methods
init()                  初始化方法        创建订单对象时初始化属性，设置订单号、租用者、单车、开始时间等。
startRent()             开始租用          记录租用开始时间，设置订单状态为进行中, 耦合车进行车辆租借
endRent()               结束租用          记录租用结束时间，计算租用费用，设置订单状态为已完成
orderDetails()          订单详情          返回订单的详细信息，包括订单号、租用者、单车、开始时间、结束时间、费用等
close()                 关闭订单          核查支付，改变是否支付状态
cancelOrder()           取消订单          取消订单，设置状态为已取消
